```python
def chi2_distance(A, B):
        
    # compute the chi-squared distance using above formula 
    chi = 0.5 * np.nansum([((a - b) ** 2) / (a + b)  
                      for (a, b) in zip(A, B)]) 
    return chi 
  
```
